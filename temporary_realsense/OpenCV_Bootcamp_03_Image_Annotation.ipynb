### __Annotating Images__

In this notebook we will cover how to annotate images using OpenCV. We will learn how to peform the following annotations to images.

- Draw lines
- Draw circles
- Draw rectangles
- Add text

These are useful when you want to annotate your results for presentations or show a demo of your application. Annotations can also be useful during development and debugging.

```python
# Import libraries
import os
import cv2
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)

%matplotlib inline
```

### __Download Assets__

```python
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
```

```python
URL = r"https://www.dropbox.com/s/48hboi1m4crv1tl/opencv_bootcamp_assets_NB3.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB3.zip")

# Download if assest ZIP does not exists. 
if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path)   
```

```python
# Read in an image
image = cv2.imread("Apollo_11_Launch.jpg", cv2.IMREAD_COLOR)

# Display the original image
plt.imshow(image[:, :, ::-1])
```

### __Drawing a Line__

Let's start off by drawing a line on an image. We will use <span style='color: black; font-family: monospace; background-color: lightgray;'>__cv2.line__</span> function for this.

<font color='green'>__Function Syntx__</font>

_img = cv2.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]])_

<span style='color: black; font-family: monospace; background-color: lightgray;'>img</span>: The output image that has been annotated.

The function has __4 required arguments__:

1. <span style='color: black; font-family: monospace; background-color: lightgray;'>img</span>: Image on which we will draw a line

2. <span style='color: black; font-family: monospace; background-color: lightgray;'>pt1</span>: First point(x,y location) of the line segment

3. <span style='color: black; font-family: monospace; background-color: lightgray;'>pt2</span>: Second point of the line segment

4. <span style='color: black; font-family: monospace; background-color: lightgray;'>color</span>: Color of the line which will be drawn

Other optional arguments that are important for us to know include:

1. <span style='color: black; font-family: monospace; background-color: lightgray;'>thickness</span>: Integer specifying the line thickness. Default value is 1.

2. <span style='color: black; font-family: monospace; background-color: lightgray;'>lineType</span>: Type of the line. Default value is 8 which stands for an 8-connected line. Usually, cv2.LINE_AA (antialiased or smooth line) is used for the lineType.

<font color='green'>__OpenCV Documentation__</font>

<span style='color: black; font-family: monospace; background-color: lightgray;'>line</span> <a href="https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#ga7078a9fae8c7e7d13d24dac2520ae4a2" target="_blank">Documentation Link</a>

Let's see an example of this.

```python
imageLine = image.copy()

# The line starts from (200,100) and ends at (400,100)
# The color of the line is YELLOW (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.line(imageLine, (200, 100), (400, 100), (0, 255, 255), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageLine[:,:,::-1])
```

### __Drawing a Circle__

Let's start off by drawing a circle on an image. We will use <span style='color: black; font-family: monospace; background-color: lightgray;'>__cv2.circle__</span> function for this.

<font color='green'>__Function Syntx__</font>

_img = cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]])_

<span style='color: black; font-family: monospace; background-color: lightgray;'>img</span>: The output image that has been annotated.

The function has __4 required arguments__:

1. <span style='color: black; font-family: monospace; background-color: lightgray;'>img</span>: Image on which we will draw a line

2. <span style='color: black; font-family: monospace; background-color: lightgray;'>center</span>: Center of the circle

3. <span style='color: black; font-family: monospace; background-color: lightgray;'>radius</span>: Radius of the circle

4. <span style='color: black; font-family: monospace; background-color: lightgray;'>color</span>: Color of the line which will be drawn

Other optional arguments that are important for us to know include:

1. <span style='color: black; font-family: monospace; background-color: lightgray;'>thickness</span>: Integer specifying the line thickness. Default value is 1.

2. <span style='color: black; font-family: monospace; background-color: lightgray;'>lineType</span>: Type of the line. Default value is 8 which stands for an 8-connected line. Usually, cv2.LINE_AA (antialiased or smooth line) is used for the lineType.

<font color='green'>__OpenCV Documentation__</font>

<span style='color: black; font-family: monospace; background-color: lightgray;'>circle</span> <a href="https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#gaf10604b069374903dbd0f0488cb43670" target="_blank">Documentation Link</a>

Let's see an example of this.

```python
# Draw a circle
imageCircle = image.copy()

cv2.circle(imageCircle, (900,500), 100, (0, 0, 255), thickness=5, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(imageCircle[:,:,::-1])
```

### __Drawing a Rectangular__

Let's start off by drawing a rectengular on an image. We will use <span style='color: black; font-family: monospace; background-color: lightgray;'>__cv2.rectangle__</span> function for this.

<font color='green'>__Function Syntx__</font>

_img = cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])_

<span style='color: black; font-family: monospace; background-color: lightgray;'>img</span>: The output image that has been annotated.

The function has __4 required arguments__:

1. <span style='color: black; font-family: monospace; background-color: lightgray;'>img</span>: Image on which we will draw a line

2. <span style='color: black; font-family: monospace; background-color: lightgray;'>pt1</span>: Vertex of the rectangle. Usually we use the __top-left vertex__ here.

3. <span style='color: black; font-family: monospace; background-color: lightgray;'>pt2</span>: Vertex of the rectangle opposite to pt1. Usually we use the __bottom-right vertex__ here.

4. <span style='color: black; font-family: monospace; background-color: lightgray;'>color</span>: Color of the line which will be drawn

Other optional arguments that are important for us to know include:

1. <span style='color: black; font-family: monospace; background-color: lightgray;'>thickness</span>: Integer specifying the line thickness. Default value is 1.

2. <span style='color: black; font-family: monospace; background-color: lightgray;'>lineType</span>: Type of the line. Default value is 8 which stands for an 8-connected line. Usually, cv2.LINE_AA (antialiased or smooth line) is used for the lineType.

<font color='green'>__OpenCV Documentation__</font>

<span style='color: black; font-family: monospace; background-color: lightgray;'>rectangle</span> <a href="https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#ga07d2f74cadcf8e305e810ce8eed13bc9" target="_blank">Documentation Link</a>

Let's see an example of this.

```python
# Draw a rectangle (thickness is a positive integer)
imageRectangle = image.copy()

cv2.rectangle(imageRectangle, (500, 100), (700, 600), (255, 0, 255), thickness=5, lineType=cv2.LINE_8)

# Display the image
plt.imshow(imageRectangle[:, :, ::-1])
```

### __Adding a Text__

Finally, let's see how we can write some text on an image using  <span style='color: black; font-family: monospace; background-color: lightgray;'>__cv2.putText__</span> function.

<font color='green'>__Function Syntx__</font>

_img = cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])_

<span style='color: black; font-family: monospace; background-color: lightgray;'>img</span>: The output image that has been annotated.

The function has __6 required arguments__:

1. <span style='color: black; font-family: monospace; background-color: lightgray;'>img</span>: Image on which we will draw a line

2. <span style='color: black; font-family: monospace; background-color: lightgray;'>text</span>: Text string to be written.

3. <span style='color: black; font-family: monospace; background-color: lightgray;'>org</span>: Bottom-left corner of the text string in the image.

4. <span style='color: black; font-family: monospace; background-color: lightgray;'>fontFace</span>: Font type

5. <span style='color: black; font-family: monospace; background-color: lightgray;'>fontScale</span>: Font scale factor that is multiplied by the font-specific base size

6. <span style='color: black; font-family: monospace; background-color: lightgray;'>color</span>: Font color

Other optional arguments that are important for us to know include:

1. <span style='color: black; font-family: monospace; background-color: lightgray;'>thickness</span>: Integer specifying the line thickness. Default value is 1.

2. <span style='color: black; font-family: monospace; background-color: lightgray;'>lineType</span>: Type of the line. Default value is 8 which stands for an 8-connected line. Usually, cv2.LINE_AA (antialiased or smooth line) is used for the lineType.

<font color='green'>__OpenCV Documentation__</font>

<span style='color: black; font-family: monospace; background-color: lightgray;'>putText</span> <a href="https://docs.opencv.org/4.5.1/d6/d6e/group__imgproc__draw.html#ga5126f47f883d730f633d74f07456c576" target="_blank">Documentation Link</a>

Let's see an example of this.

```python
imageText = image.copy()
text = "Apollo 11 Saturn V Launch, July 16, 1969"
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 0)
fontThickness = 2

cv2.putText(imageText, text, (200, 700), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA);

# Display the image
plt.imshow(imageText[:, :, ::-1])
```
